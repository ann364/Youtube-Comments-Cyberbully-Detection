##### Adding frequent words to stop words

```python
# Let's aggregate this list and identify the most common words along with how many routines they occur in
Counter(words).most_common()

# If more than half of the video have it as a top word, exclude it from the list
add_stop_words = [word for word, count in Counter(words).most_common() if count > 4]
add_stop_words
```

```python
def multiple_appends(listname, *element):
    listname.extend(element)

multiple_appends(add_stop_words, "dani", "danielle", "madison","rebecca","shannon","really","richard",'casey','emily','steven','shaheen')
print(add_stop_words)
```

```python
# Let's update our document-term matrix with the new list of stop words

# Read in cleaned data
personComments_corpus= pd.read_pickle('personComments_corpus.pkl')
# personComments = personComments.rename(columns={'Cleaned_Comments':'comments'})

# Add new stop words
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

# Recreate document-term matrix
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(personComments_corpus.comments)
personComments_dtm2 = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
personComments_dtm2.index = personComments_corpus.index
personComments_dtm2
```

```python
# Pickle it for later use
pickle.dump(cv, open("cv_stop.pkl", "wb"))
personComments_dtm2.to_pickle("personComments_dtm2.pkl")
```

### Unique Words

```python
# Find the number of unique words in comments for each uploaders

# Identify the non-zero items in the document-term matrix, meaning that the word occurs at least once
unique_list = []
personComments_dtm.columns
for person in personComments_dtm.columns:
    uniques = personComments_dtm[person].to_numpy().nonzero()[0].size
    unique_list.append(uniques)

# Create a new dataframe that contains this unique word count
person_uniqueWords = pd.DataFrame(list(zip(person_list, unique_list)), columns=['Person', 'unique_words'])
person_uniqueWords_sorted = person_uniqueWords.sort_values(by='unique_words')
```

```python
y=person_uniqueWords_sorted.unique_words
x= person_uniqueWords_sorted.Person
fig, ax = plt.subplots()    
y_pos = np.arange(len(person_uniqueWords))
ax.barh(y_pos, y)
plt.yticks(y_pos, person_uniqueWords_sorted.Person)
plt.xlabel("Count- Unique Words", fontsize=15)
plt.ylabel("People", fontsize=15)
plt.title('Number of Unique Words in comments for each person', fontsize=20) 

for i, v in enumerate(y):
    ax.text(v + 3, i + .25, str(v), color='black', fontweight='bold')
    
plt.show()
```

### Profanity

```python
# Earlier I said we'd revisit profanity. Let's take a look at the most common words again.
Counter(words).most_common()

# Let's isolate just these bad words
data_bad_words = personComments_dtm.transpose() [['fucking','shit','gay','wtf','stupid','ass','fat','fuck',
                                                           'huge','friday','hate','looks','bullied','lesbian','bully','arabic']]
data_bad_words
data_profanity = pd.concat([data_bad_words.fucking+data_bad_words.wtf+data_bad_words.shit+data_bad_words.gay+data_bad_words.stupid+data_bad_words.ass+data_bad_words.fat+data_bad_words.fuck,
                            data_bad_words.friday+data_bad_words.huge+data_bad_words.hate+data_bad_words.looks+data_bad_words.bullied+data_bad_words.lesbian+data_bad_words.bully+data_bad_words.arabic],
                           axis=1)
data_profanity.columns = ['swear_words', 'milder_words']
data_profanity

# scatter_x = data_profanity.swear_words.to_list()
# scatter_y = data_profanity.milder_words.to_list()
# group = data_profanity.index.to_list()

# plt.scatter(scatter_x, scatter_y, label=group)
# plt.legend()

# Let's create a scatter plot of our findings
plt.rcParams['figure.figsize'] = [10, 8]

for i, person_list in enumerate(data_profanity.index):
    x = data_profanity.swear_words.loc[person_list]
    y = data_profanity.milder_words.loc[person_list]
    plt.scatter(x, y, color='blue')
    plt.text(x=x+1.5,y=y+1.5,s=person_list, fontsize=10)

plt.title('Bad Words Used ', fontsize=20)
plt.xlabel('Number of Swear Words', fontsize=15)
plt.ylabel('Number of Milder Words', fontsize=15)

plt.show()
```
